```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score, confusion_matrix, classification_report
import seaborn as sns
```

```python
df_diamonds = sns.load_dataset('diamonds')
df_diamonds.head()
```

```python
x = df_diamonds['carat'].values.reshape(-1, 1)
y_continuous = df_diamonds['price'].values.reshape(-1, 1)
```

```python
# Crear una variable binaria para clasificación
threshold = 5000
y_classification = (y_continuous > threshold).astype(int).ravel()
```

```python
# División única y coherente
x_train, x_test, y_train_continuous, y_test_continuous, y_train_classification, y_test_classification = train_test_split(
    x, y_continuous, y_classification, test_size=0.2, random_state=42
)
```

```python
# Modelo de regresión lineal
linear_model = LinearRegression()
linear_model.fit(x_train, y_train_continuous)
y_pred_continuous = linear_model.predict(x_test)

# Evaluación de regresión lineal
mse = mean_squared_error(y_test_continuous, y_pred_continuous)
print(f"Error cuadrático medio (MSE): {mse:.2f}")
```

```python
# Modelo de regresión logística
logistic_model = LogisticRegression()
logistic_model.fit(y_pred_continuous.reshape(-1, 1), y_test_classification)
y_pred_classification = logistic_model.predict(y_pred_continuous.reshape(-1, 1))

# Evaluación de regresión logística
accuracy = accuracy_score(y_test_classification, y_pred_classification)
print(f"Exactitud (Accuracy): {accuracy:.2f}")
print("Matriz de confusión:")
print(confusion_matrix(y_test_classification, y_pred_classification))
print("Informe de clasificación:")
print(classification_report(y_test_classification, y_pred_classification))
```

```python
# Visualización
plt.figure(figsize=(12, 6))

# Regresión lineal
plt.subplot(1, 2, 1)
plt.scatter(x_test, y_test_continuous, color='blue', label='Precio real')
plt.plot(x_test, y_pred_continuous, color='red', label='Precio estimado')
plt.xlabel('Carat')
plt.ylabel('Precio')
plt.title('Regresión Lineal')
plt.legend()

# Clasificación
plt.subplot(1, 2, 2)
plt.scatter(y_pred_continuous, y_test_classification, color='blue', label='Precio real')
plt.scatter(y_pred_continuous, y_pred_classification, color='red', marker='x', label='Precio estimado')
plt.xlabel('Carat')
plt.ylabel('Precio')
plt.title('Regresión Logística')
plt.legend()

plt.tight_layout()
plt.show()
```
