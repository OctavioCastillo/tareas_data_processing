```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
df_diabetes = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
df_diabetes['progression'] = diabetes.target
df_diabetes.head()
```

```python
df_diabetes.isna().sum()
#df_diabetes.describe()
```

```python
# Una variable independiente
x = df_diabetes['bmi'].values.reshape(-1, 1)
y = df_diabetes['progression'].values.reshape(-1, 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)

regressor = LinearRegression()
regressor.fit(x_train, y_train)
regressor.score(x_test, y_test)
```

```python
y_pred = regressor.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print('Coefficients: \n', regressor.coef_)
print('Independent term: \n', regressor.intercept_)
print("Mean squared error: %.2f" % mean_squared_error(y_train, y_pred))
```

```python
sns.set_theme()
flg, ax = plt.subplots(1, 2, figsize=(10, 10), sharey=True)
ax[0].scatter(x_train, y_train)
ax[0].plot(x_train, regressor.predict(x_train), c='g')
ax[0].set_title('bmi vs progression (trainin)')
ax[0].set_xlabel('bmi')
ax[0].set_ylabel('progression')
ax[1].scatter(x_test, y_test)
ax[1].plot(x_train, regressor.predict(x_train), c='g')
ax[1].set_title('bmi vs progression (testing)')
ax[1].set_xlabel('bmi')
ax[1].set_ylabel('progression')
plt.suptitle('Linear Regression Model')
```

```python
# Dos variables independientes
xx = df_diabetes[['bmi', 's6']].values
yy = df_diabetes['progression'].values.reshape(-1, 1)
xx_train, xx_test, yy_train, yy_test = train_test_split(xx, yy, test_size=0.3, random_state=0)

regressor = LinearRegression()
regressor.fit(xx_train, yy_train)
regressor.score(xx_test, yy_test)
```

```python
yy_pred = regressor.predict(xx_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print('Coefficients: \n', regressor.coef_)
print('Independent term: \n', regressor.intercept_)
print("Mean squared error: %.2f" % mean_squared_error(yy_train, yy_pred))
```
