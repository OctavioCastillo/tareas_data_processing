```python
# imports
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import cm
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# cargar datos con pandas
data = pd.read_csv("./articulos_ml.csv")
#registros y dimensiones
data.shape
```

```python
# 161 registros 8 columnas
data.head()
```

```python
#estadísticas
data.describe()
```

```python
# Visualizamos rápidamente las características de entrada
data.drop(['Title','url', 'Elapsed days'],axis=1).hist()
plt.show()
```

```python
# eje x entre 0 y 3,500
# eje y entre 0 y 80,000
filtered_data = data[(data['Word count'] <= 3500) & (data['# Shares'] <= 80000)]

colores = ['orange', 'blue']
tamanios=[30,60]

f1 = filtered_data['Word count'].values
f2 = filtered_data['# Shares'].values

# marcar en colores los puntos por encima de la media
asignar=[]
for index, row in filtered_data.iterrows():
    if(row['Word count']>1808):
        asignar.append(colores[0])
    else:
        asignar.append(colores[1])

plt.scatter(f1, f2, c=asignar, s=tamanios[0])
plt.show()
```

```python
# asignación de la variable x
dataX = filtered_data[["Word count"]]
x_train = np.array(dataX)
y_train = filtered_data['# Shares'].values

# objeto regresión lineal
regr = linear_model.LinearRegression()

# Entrenamos nuestro modelo
regr.fit(x_train, y_train)

# hacemos las predicciones que en definitica una linea (en este caso, al ser 20)
y_pred = regr.predict(x_train)

# veamos los coeficientees obtenidos. En este caso son la tangente
print('Coefficients: \n', regr.coef_)
# este es el valor donde corta el eje y (en x=0)
print("Mean squared error: %.2f" % mean_squared_error(y_train, y_pred))
# puntaje de varianza. El mejor puntaje es un 1.0
print('Variance score: %.2f' % r2_score(y_train, y_pred))
```

```python
plt.scatter(f1, f2, c=asignar, s=tamanios[0])
#plt.show()
plt.plot(f1, y_pred, color='red', linewidth=3)
plt.xlabel('Word count')
plt.ylabel('# Shares')
plt.title('Gráfico de dispersión con la línea de regresión')
plt.show()
```

```python
# comprobación
# quiero predecir cuántos "Shares" voy a obtener por un articulo con 2.000 palabras
# según nuestro modelo, hacemos:
y_Dosmil = regr.predict([[2000]])
print(int(y_Dosmil[0]))
```
